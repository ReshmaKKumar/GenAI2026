```python
from sentence_transformers import SentenceTransformer, util

# Load a pretrained model (correct model name)
model = SentenceTransformer('all-MiniLM-L6-v2')

# Sentences to encode
sentences = [
    "I love pizzza",
    "pizza is my favourite food",
    "stock market is up today"
]

# Convert sentences to embeddings
embeddings = model.encode(sentences)
print("Shape :" , embeddings.shape)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
scores = cosine_similarity(embeddings)
print("\nsimilarity scores:")
for i in range(len(sentences)):
  for j in range(i+1,len(sentences)):
    print(f" '{sentences[i]}'")
    print(f" '{sentences[j]}'")
    print(f"  score -> {scores[i][j]:.2f}\n")
```

```python
import chromadb

# Create client
client = chromadb.PersistentClient(path="./my_vector_db")

# Create collection
collection = client.get_or_create_collection(name="my_docs")

# Add documents
collection.add(
    documents=[
        "Python is a programming language used for AI",
        "Machine learning is a subset of artificial intelligence",
        "Cricket world cup final results in 2024",
        "Deep learning uses neural networks",
    ],
    metadatas=[
        {"topic": "tech"},
        {"topic": "ai"},
        {"topic": "sports"},
        {"topic": "ai"},
    ],
    ids=["1", "2", "3", "4"]  # REQUIRED (important!)
)
```

```python
results = collection.query(
    query_texts=["What is artificial intelligence?"],  # your question
    n_results=2
)

print(results)
```
